```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
file_path = r"C:\Users\Kento\Documents\Exeter\python\dataset for the blog\share-of-adults-who-smoke.csv"

# 
column_names = ['Entity', 'Code', 'Year', 'Prevalence of tobacco use (% of adults)']

# CSV
df = pd.read_csv(file_path, names=column_names)

# 'Prevalence of tobacco use (% of adults)' 
df['Prevalence of tobacco use (% of adults)'] = pd.to_numeric(df['Prevalence of tobacco use (% of adults)'], errors='coerce')

# 
major_countries = ['United States', 'United Kingdom', 'France', 'Japan']

# 
plt.figure(figsize=(10, 6))

# 
min_value = df[df['Entity'].isin(major_countries)]['Prevalence of tobacco use (% of adults)'].min()
max_value = df[df['Entity'].isin(major_countries)]['Prevalence of tobacco use (% of adults)'].max()

# 
for country in major_countries:
    country_data = df[df['Entity'] == country]
    plt.plot(country_data['Year'], country_data['Prevalence of tobacco use (% of adults)'], marker='o', label=country)

# 
plt.ylim(min_value - 5, max_value + 5)

# 
plt.yticks(range(int(min_value) - 5, int(max_value) + 6, 5))

# 
plt.title('Prevalence of tobacco use')
plt.xlabel('Year')
plt.ylabel('Prevalence of tobacco use (% of adults)')

# 
plt.legend()

# 
plt.grid(True, axis='y', linestyle='--')

# 
plt.tight_layout()
plt.savefig('share-of-adults-who-smoke.png')
plt.show()
```
